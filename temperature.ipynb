```python
import pandas as pd

df = pd.read_csv('azureml://subscriptions/b43bf58b-22dc-4419-ae40-26f681e3cd05/resourcegroups/t-aml-1/workspaces/t-aml-ws-1/datastores/workspaceblobstore/paths/UI/2025-05-07_065026_UTC/temperature.csv')

df
```

```python
# Reformat date column to date time. The previous format was 'dd/mm/yyyy hh:mm'
df['Date'] =  pd.to_datetime(df['Date'], format='%d/%m/%Y %H:%M')

# Interpolate missing values in Vancouver column
df['Vancouver'] = df['Vancouver'].interpolate(method='linear')

# Set first row value by copying the second row value
df.loc[0, 'Vancouver'] = df.loc[1, 'Vancouver']

# Pick only Date and Vancouver columns and save to CSV
df[['Date', 'Vancouver']].to_csv('temperature.csv', index=False)
```
